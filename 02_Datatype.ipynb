# Python Built-in Data Types

Python has a few built-in data types. Different sources may classify
them slightly differently, but a commonly accepted list includes:

-   Numeric
-   Sequence
-   Set
-   Dictionary
-   Boolean

Some of these types also contain sub-types.

## Numeric

Numeric data types represent numbers. Python supports:

-   **int** (Integer): Whole numbers such as `-10`, `0`, `42`,
    `8675309`.

    -   Underscores can be used as digit separators:

        ``` python
        x = 867_5309
        print(x)
        # 8675309
        ```

-   **float**: Numbers with decimal points or scientific notation.

    ``` python
    x = 10e100
    print(x)   # 1e+100

    x = 10e400
    print(x)   # inf
    ```

-   **complex**: Numbers containing a real and imaginary part.

    ``` python
    a = 10 + 3j
    print(a)
    # (10+3j)
    ```

## Sequence

Sequence types are ordered collections.

Examples include:

-   **str** (String)
-   **list**
-   **tuple**

Example:

``` python
x = "Hello, New Stack"
print(x)
```

Output:

    Hello, New Stack

Lists use `[]` and are mutable, while tuples use `()` and are immutable.

## Set

A set is:

-   Unordered
-   Iterable
-   Mutable
-   Contains only unique elements

Example:

``` python
x = {1, 2, 3, 4, 5, "New Stack"}
print(x)
```

Union example:

``` python
x = {"Hello"}
y = {"New Stack"}

print(x | y)
```

Output:

    {'Hello', 'New Stack'}

## Dictionary

A dictionary stores data as key-value pairs.

Example:

``` python
data = {
    "name": "Jack Wallen",
    "client": "The New Stack",
    "subject": "Python"
}

print(data)
print(data["client"])
```

Output:

    {'name': 'Jack Wallen', 'client': 'The New Stack', 'subject': 'Python'}

    The New Stack

## Boolean

A Boolean value is either `True` or `False`.

Examples:

``` python
print(10 > 5)
# True

print(5 > 10)
# False
```

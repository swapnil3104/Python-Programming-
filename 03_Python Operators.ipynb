# Python Operators

Operators are used to perform operations on variables and values.

## Python Arithmetic Operators

Arithmetic operators are used with numeric values to perform common mathematical operations such as addition, subtraction, multiplication, and division.

---

### Arithmetic Operators Table

| Operator | Name | Description | Example |
|----------|------|-------------|---------|
| `+` | Addition | Adds two numbers | `x + y` |
| `-` | Subtraction | Subtracts one number from another | `x - y` |
| `*` | Multiplication | Multiplies two numbers | `x * y` |
| `/` | Division | Divides one number by another and returns a float | `x / y` |
| `%` | Modulus | Returns the remainder after division | `x % y` |
| `**` | Exponentiation | Raises a number to the power of another | `x ** y` |
| `//` | Floor Division | Returns the quotient without the decimal part | `x // y` |

---


```python
x = 10
y = 3

print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Modulus:", x % y)
print("Exponentiation:", x ** y)
print("Floor Division:", x // y)
```

### Division in Python

Python has two division operators:

    / - Division (returns a float)
    // - Floor division (returns an integer)


```python
# Division always returns a float:
x = 12
y = 5

print(x / y)
```

```python
# Floor division always returns an integer.
# It rounds DOWN to the nearest integer:

x = 12
y = 5

print(x // y)
```

## Python Assignment Operators

Assignment operators are used to assign values to variables. They can also perform an operation and assign the result back to the same variable.

---

### Assignment Operators Table

| Operator | Example | Equivalent Expression | Description |
|----------|---------|-----------------------|-------------|
| `=` | `x = 5` | `x = 5` | Assigns a value to a variable |

| `-=` | `x -= 3` | `x = x - 3` | Subtracts and assigns |
| `*=` | `x *= 3` | `x = x * 3` | Multiplies and assigns |
| `/=` | `x /= 3` | `x = x / 3` | Divides and assigns |
| `%=` | `x %= 3` | `x = x % 3` | Modulus and assigns |
| `//=` | `x //= 3` | `x = x // 3` | Floor divides and assigns |
| `**=` | `x **= 3` | `x = x ** 3` | Exponentiates and assigns |
| `&=` | `x &= 3` | `x = x & 3` | Bitwise AND and assigns |

### The Ternary Operator

The ternary operator allows you to assign one value if a condition is true, and another if it is false:

Example

Assign a value to x:

Assign the value "WEEKEND!" if the number is higher than 5, otherwise "Workday":

```python
num = 6

x = "WEEKEND!" if num > 5 else "Workday"

print(x)
```

Instead of Elif:

The ternary operator can be used instead of elif in longer if statements:

Example

Assign:
- "Fri" if num is 5
- "Sat" if num is 6
- "Sun" if num is 7
- otherwise assign "weekday":

```python
num = 6

x = "Fri" if num == 5 else "Sat" if num == 6 else "Sun" if num == 7 else "weekday"

print(x)
```

## Python Comparison Operators

Comparison operators are used to compare two values. They return a Boolean value:

- `True` if the comparison is correct.
- `False` if the comparison is incorrect.

These operators are commonly used in **conditional statements**, **loops**, and **decision-making** in Python.

---

### Comparison Operators Table

| Operator | Name | Description | Example |
|----------|------|-------------|---------|
| `==` | Equal | Returns `True` if both values are equal | `x == y` |
| `!=` | Not Equal | Returns `True` if values are not equal | `x != y` |
| `>` | Greater Than | Returns `True` if the left value is greater | `x > y` |
| `<` | Less Than | Returns `True` if the left value is smaller | `x < y` |
| `>=` | Greater Than or Equal To | Returns `True` if the left value is greater than or equal to the right value | `x >= y` |
| `<=` | Less Than or Equal To | Returns `True` if the left value is less than or equal to the right value | `x <= y` |

---

```python
x = 5
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

### Chaining Comparison Operators

Python allows you to chain comparison operators:

```python
x = 5

print(1 < x < 10)

print(1 < x and x < 10)
```

## Python Logical Operators

Logical operators are used to combine or modify conditional statements. They return a Boolean value (`True` or `False`) and are commonly used in **if statements**, **loops**, and other decision-making structures.

Python provides three logical operators:

- `and`
- `or`
- `not`

---

### Logical Operators Table

| Operator | Description | Example |
|----------|-------------|---------|
| `and` | Returns `True` if both conditions are true | `x < 5 and x < 10` |
| `or` | Returns `True` if at least one condition is true | `x < 5 or x < 4` |
| `not` | Reverses the result of a condition | `not(x < 5 and x < 10)` |

---

Example

Test if a number is greater than 0 and less than 10:

```python
x = 5

print(x > 0 and x < 10)
```

```python
# Test if a number is less than 5 or greater than 10: 
x = 5

print(x < 5 or x > 10)
```

```python
# Reverse the result with not:
x = 5

print(not(x > 3 and x < 10))
```

```python
print(5 < 7 or 5 > 7) 
```

## Python Identity Operators

Identity operators are used to compare **whether two variables refer to the same object in memory**, not whether their values are equal.

Python provides two identity operators:

- `is`
- `is not`

> **Note:** Identity operators compare the **memory location (object identity)**, while comparison operators (such as `==`) compare the **values**.

---

### Identity Operators Table

| Operator | Description | Example |
|----------|-------------|---------|
| `is` | Returns `True` if both variables refer to the same object | `x is y` |
| `is not` | Returns `True` if both variables do not refer to the same object | `x is not y` |

---


Examples

```python
# The is operator returns True if both variables point to the same object:
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z) # True, because z is the same object as x.
print(x is  y) # False, because x and y are not the same object, even if they have the same content.
print(x is not y) # True, because x is not the same object as y.
print(x == y) # True, because x is equal to y.
```

```python
# The is not operator returns True if both variables do not point to the same object:
x = ["apple", "banana"]
y = ["apple", "banana"]

print(x is not y) # True, because x is not the same object as y, even if they have the same content.
```

Difference Between is and ==

    is - Checks if both variables point to the same object in memory
    == - Checks if the values of both variables are equal


```python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y) # True, because x and y have the same content.
print(x is y) # False, because x and y are not the same object, even if they have the same content.
```

## Python Membership Operators

Membership operators are used to test whether a value or sequence is present in an object such as a **string**, **list**, **tuple**, **set**, or **dictionary**.

Python provides two membership operators:

- `in`
- `not in`

These operators return a Boolean value:

- `True` if the condition is satisfied.
- `False` otherwise.

---

### Membership Operators Table

| Operator | Description | Example |
|----------|-------------|---------|
| `in` | Returns `True` if the specified value is present in the object | `x in y` |
| `not in` | Returns `True` if the specified value is not present in the object | `x not in y` |

---

```python
# Check if "banana" is present in a list:
fruits = ["apple", "banana", "cherry"]

print("banana" in fruits) # True, because "banana" is present in the list.
```

```python
# Check if "pineapple" is NOT present in a list:
fruits = ["apple", "banana", "cherry"]

print("pineapple" not in fruits) # True, because "pineapple" is not present in the list.
```

Membership in Strings

The membership operators also work with strings:



```python
text = "Hello World"

print("H" in text) # True, because "H" is present in the string.
print("hello" in text) # False, because "hello" is not present in the string.
print("z" not in text) # True, because "z" is not present in the string.
```

## Python Bitwise Operators

Bitwise operators are used to perform operations on the binary (bit-level) representation of integers. These operators compare or manipulate individual bits of numbers.

Python provides the following bitwise operators:

- `&` (AND)
- `|` (OR)
- `^` (XOR)
- `~` (NOT)
- `<<` (Left Shift)
- `>>` (Right Shift)

---

### Bitwise Operators Table

| Operator | Name | Description | Example |
|----------|------|-------------|---------|
| `&` | AND | Sets each bit to `1` if both bits are `1` | `x & y` |
| `\|` | OR | Sets each bit to `1` if at least one bit is `1` | `x \| y` |
| `^` | XOR | Sets each bit to `1` if only one bit is `1` | `x ^ y` |
| `~` | NOT | Inverts all the bits | `~x` |
| `<<` | Left Shift | Shifts bits to the left by adding zeros on the right | `x << 2` |
| `>>` | Right Shift | Shifts bits to the right by discarding the rightmost bits | `x >> 2` |

---

```python
x = 10
y = 4

print("AND :", x & y)
print("OR  :", x | y)
print("XOR :", x ^ y)
print("NOT :", ~x)
print("Left Shift :", x << 2)
print("Right Shift:", x >> 2)
```

## Python Operator Precedence

Operator precedence determines the order in which Python evaluates operators in an expression. Operators with **higher precedence** are evaluated before operators with **lower precedence**.

If two operators have the **same precedence**, Python evaluates them **from left to right** (except for exponentiation `**`, which is evaluated from right to left).

---

### Operator Precedence Table

The following table lists Python operators from **highest precedence** to **lowest precedence**.

| Precedence | Operator(s) | Description |
|------------|-------------|-------------|
| 1 (Highest) | `()` | Parentheses |
| 2 | `**` | Exponentiation |
| 3 | `+x`, `-x`, `~x` | Unary plus, unary minus, bitwise NOT |
| 4 | `*`, `/`, `//`, `%` | Multiplication, division, floor division, modulus |
| 5 | `+`, `-` | Addition and subtraction |
| 6 | `<<`, `>>` | Bitwise left shift, right shift |
| 7 | `&` | Bitwise AND |
| 8 | `^` | Bitwise XOR |
| 9 | `\|` | Bitwise OR |
| 10 | `==`, `!=`, `>`, `>=`, `<`, `<=`, `is`, `is not`, `in`, `not in` | Comparison, identity, and membership operators |
| 11 | `not` | Logical NOT |
| 12 | `and` | Logical AND |
| 13 (Lowest) | `or` | Logical OR |

---


```python

# Example
# Get your own Python Server
# Parentheses has the highest precedence, meaning that expressions inside parentheses must be evaluated first:

print((6 + 3) - (6 + 3))

```

- Parentheses `()` have the highest precedence.
- Multiplication and division are evaluated before addition and subtraction.
- Comparison operators are evaluated after arithmetic operators.
- Logical operators are evaluated last.
- Operators with the same precedence are evaluated from **left to right**, except for exponentiation (`**`), which is evaluated from **right to left**.
- Use parentheses to make expressions easier to read and avoid ambiguity.
